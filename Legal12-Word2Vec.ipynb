#### Function for preprocessing

```python
def preprocess_and_write_to_file(directory_path, output_file_path):
    encodings = ['utf-8', 'latin1']
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        for subdir, dirs, files in os.walk(directory_path):
            for file in files:
                if file.endswith('.txt'):
                    file_path = os.path.join(subdir, file)
                    for encoding in encodings:
                        try:
                            with open(file_path, 'r', encoding=encoding) as input_file:
                                text = input_file.read()
                                preprocessed_text = simple_preprocess(text)
                                output_file.write(" ".join(preprocessed_text) + "\n")
                                break
                        except UnicodeDecodeError:
                            pass
                    else:
                        print(f"Failed to decode file: {file_path}")
```

```python
preprocess_and_write_to_file(folder_path, output_file)
```

```python
class MyCorpus:
    def __iter__(self):
        with open(output_file, 'r', encoding='utf-8') as f:
            for line in f:
                yield simple_preprocess(line)

corpus = MyCorpus()
model = Word2Vec(corpus, vector_size=100, window=5, min_count=5, workers=4)

model.save(model_save_path)
```

```python
def calculate_cosine_similarity(model, doc1, doc2):
    doc1_processed = preprocess_text(doc1)
    doc2_processed = preprocess_text(doc2)
    
    doc1_filtered = [word for word in doc1_processed if word in model.wv.key_to_index]
    doc2_filtered = [word for word in doc2_processed if word in model.wv.key_to_index]
    
    
    if doc1_filtered and doc2_filtered:
        vec1 = model.wv[doc1_filtered]
        vec2 = model.wv[doc2_filtered]
        similarity_matrix = cosine_similarity(vec1, vec2)
        return similarity_matrix
    else:
        return None
```

```python

def read_text_from_file(file_path):
    encodings = ['utf-8', 'latin-1']
    for encoding in encodings:
        try:
            with open(file_path, 'r', encoding=encoding) as file:
                return file.read()
        except UnicodeDecodeError:
            continue
    return None
```

```python
def find_file_path(file_name):
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file == file_name:
                return os.path.join(root, file)
    return None
```
